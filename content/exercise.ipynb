## Example 2

You collected experimental measurements. Now you are looking at their average.

````python

measurements = [2.1, 8.7, 2.3, 4.5, 2.9, 3.0, 999.0, 2.8, 2.5, 2.6, 2.4, 1.3]

mean_measurements = sum(measurements) / len(measurements)

print(mean_measurements)
````

One value looks suspiciously high, so we want to test what happens if we replace it with a more plausible value. Will we get a more reasonable average?


````python
test_measurements = measurements
test_measurements[6] = 3.0
mean_test_measurements = sum(test_measurements) / len(test_measurements)
print(mean_test_measurements)
````

That looks better! Now we want to keep working with the original measurements. But hey, they changed!

````python
print(measurements)
print(mean_measurements)
````

Your goal is to test fix the bug that changes the original measurements.
